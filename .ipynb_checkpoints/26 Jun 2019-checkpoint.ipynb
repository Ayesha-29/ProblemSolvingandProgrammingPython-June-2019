```python
filename='Datafiles/contacts.txt'
import re
def phnovalidator(no):
    pattern='[9876][0-9]{9}$|^[0][6-9][0-9]{9}$|^[+][9][1][6-9][0-9]{9}$'
    if re.match(pattern,str(no)):
        return True
        #print("valid number")
    else:
        return False
        #print("invalid number")
    return
def emailval(email):
    pattern='^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][0-9a-z]{3,18}[.][a-z]{2,4}$'
    if re.match(pattern,email):
        #print("valid")
        return True
    else:
        #print("invalid")
        return False
def csvtolist(filename):
    li=[]
    with open(filename,'r') as f:
        for line in f:
            li.append(line.split(','))
    return li


def listofile(li):
    s=''
    for i in li:
        s+=','.join(i)
    return(s)
    

def addcontact(name,no,email):
    with open(filename,'a') as f:
        if not searchcontact(name):
            if phnovalidator(int(no)):
                if emailval(email):
                    f.write(name+','+str(no)+','+email+'\n')
                    print("contact added successfully")
        else:
            print("invalid,")

def searchcontact(name):
    fh=csvtolist(filename)
    for line in fh:
        
        if line[0]==name:
            
            #print(line[0],":",line[1],",",line[2])
            print("contact exists")
            return True
    print("contact does not exists" )    
    return False

def searchwithreturn(name):
    fh=csvtolist(filename)
    for line in range(len(fh)):
        if fh[line][0]==name:
            #print(line[0],":",line[1],",",line[2])
            #print("contact exists")
            return line
    return -1

def delcon(name):
    a=searchwithreturn(name)
    print(a)
    if a!=-1:
        fh=csvtolist(filename)
        fh.pop(a)
        fh=listofile(fh)
        with open(filename,'w') as f:
            f.write(fh)
        print("deleted contact  successfully")
    else:
        print("contact not found")

def listallcontacts():
    with open(filename,'r') as f:
        f=csvtolist(filename)
        for line in f:
            print(line[0])
            print("phoneno:",line[1],end=" ")
            print()
            print("email-id:",line[2],end=" ")
#listallcontacts()
def modifycontact(name,no,email):
    with open(filename,'r') as f:
        i=searchwithreturn(name)
        if i!=-1:
            fh=csvtolist(filename)
            
            for i in range(len(fh)):
                if fh[i][0]==name:
                    fh[i][1]=str(no)
                    fh[i][2]=email
                fa=listofile(fh)
                with open(filename,'w') as f:
                    f.write(fa)
            print("modified successfully")
        else:
            print("contact does not exists to modify")
                    

while 1:
    print("1.addcontact\n2.searchcontact\n3.deletecontact\n4.listallcontacts\n5.modifycontact")

    n=int(input("enter your choice"))
    
    if n==1:
        
        name,no,email=input("enter the name,no,email").split(',')
        addcontact(name,no,email)
    elif n==2:
        name=input("enter the name")
        searchcontact(name)
    elif n==3:
        name=input("enter the name")
        delcon(name)
    elif n==4:
        listallcontacts()
    elif n==5:
        name,no,email=input("enter the name,no,email").split(',')
        modifycontact(name,no,email)
    
    print("\n press 1 to continue\n0 to exit")
    p=int(input())
    if p==0:
        break
    elif p==1:
        continue
```

```python
#Function to find the second largest number

def secondlargest(li,k):
    #convert the list into a unique list
    #O(n)
    unique=[]
    if(len(li)>1):
        for n in li:
            if n not in unique:
                unique.append(n)
        unique=sorted(unique,reverse=True)
        return unique[1]
    else:
        return -1
secondlargest([1,2,3,4,2,3],2)


#def fifthleast(li):

def klargest(li,k):
    unique=[]
    
    for i in li:
        if i not in unique:
            unique.append(i)
    #sort the unique list in ascending order
    print(unique)
    #check if length of unique list is greater than k
    if(len(unique)>= k):
        return unique[k-1]
    return -1
klargest([3,3,9,6,5,4,3,5],5)

# def ksmallest(li,k):
#     #extract unique elements in the list
#     unique=[]
    
#     for i in li:
#         if i not in unique:
#             unique.append(i)
#     #sort the unique list in ascending order
#     unique=sorted(unique)
#     #check if length of unique list is greater than k
#     if(len(unique)>= k):
#         return unique[k-1]
#     return -1


# ksmallest([3,3,9],2)
```

```python
#Function to identify the element with highest frequency in a list

#If many elements have the highest frequency ,return the smallest
def  highestfrequencyelement(li,k):
    #Extracting the unique elements from
    unique={}
    for n in li:
        if n not in unique.keys():
            unique[n] = 1
        else:
            unique[n]+=1
            
            
     #unique={1:2,2:2,3:2,9:1,8:1,7:1,4:1}
     #Getting all frequencies into a list
    
    
    freq=sorted(unique.values())
    maxfreq=max(freq)
    maxfreqE=[]

    #Identify the keys with maximum frequency in a list
    
    for item in unique.items():
        if item[1]==maxfreq:
            maxfreqE.append(item[0])
    
    #select the minimum of the keys with max elements
    
    maxfreqE=(sorted(maxfreqE,reverse=True))           #minfreqE[k-1]
    return maxfreqE[k-1]
    

highestfrequencyelement([1,2,3,9,8,7,3,4,2,1],2)

```

```python


def khighestfrequency(s,k):
    #unique characters with freq in a dictionary
    unique={}
    for i in s:
        if i not in unique:
            unique[i]=1
        else:
            unique[i]+=1  
    #{9: 2, 7: 6, 6: 4, 5: 1, 2: 1, 3: 1, 4: 1}    
    #Extract all unique frequencies in a list
    l=[]
    for value in unique.values():
        if value not in l:
            l.append(value)
        
    l=sorted(l,reverse=True)
    if len(l)>k:
        kfreq=l[k-1]
    else:
        return -1
    kfreqchar=[]
    for key in unique.items():
        if key[1]==kfreq:
            kfreqchar.append(key[0])
    return min(kfreqchar)
with open('DataFiles/k-largest.txt','r') as f:
    t=int(f.readline())
    for i in range(t):
        s=f.readline()
        k=int(f.readline())
        print(khighestfrequency(s,k))


#Function to identify kth smallest frequency element
#if there are many such elements,return the smallest value
#[9,7,6,5,2,3,4,9,6,7,7,7,7,6,7,6], k=4 ---->Answer(2)
```
