```python
# Linear Graph

import matplotlib.pyplot as plt
plt.title("Basic Linear Graph")
plt.plot([1,3,5],[1,3,5])
plt.show
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
Time=list(range(1,6,1))
#d = 2 * t

plt.plot(Time,distance,'mo')
plt.title("Distance vs Time")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Distance")
plt.show()
```

```python
# Non linear graph

# v= a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
#Numbers vs Cubes
n=list(range(10))
sqn=[i**10 for i in n]
plt.plot(n,sqn)
plt.grid()
plt.show()
```

```python
# Voltage - Time
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.grid()
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]
incomearkansas=df.values[3,2:]
plt.title("Income of Arkansas")
plt.xlabel("Year")
plt.ylabel("Income")
plt.plot(years,incomearkansas)
plt.show()
```
